```
#importing pandas and numpy libraries
#
import pandas as pd

import numpy as np
```

**1.1**.**LOADING** **DATASET**

```
#Loading dataset
#
data_frame1=pd.read_csv("cells_geo.csv")
data_frame1
```

**1.2**.**CLEANING** **DATAFRAME**

```
#splitting column
#
data_frame1=data_frame1[';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE'].apply(lambda y: pd.Series(y.split(';')))
data_frame1.head()
```

```
#naming the columns
#
data_frame1.columns=['INDEX','VILLES','STATUS','LOCALISATION','DECOUPZONE','ZONENAME','LONGITUDE','LATITUDE','REGION','AREA','CELL_ID','SITE_ID']
data_frame1
```

```
#removing quotes from values
#
for i,col in enumerate(data_frame1.columns):
  data_frame1.iloc[:, i]=data_frame1.iloc[:,i].str.replace('"','')
data_frame1
```

```
#checking for any null values
#
data_frame1.info()

```

```
#checking for any duplicate values
#
data_frame1.duplicated()
```

```
#Dropping the duplicate values
#
data_frame1.drop_duplicates(inplace=True)
data_frame1
```

**2.1**.**LOADING** **TELCOM** **DATASETS**

```
#Loading first dataset
#
data_frame2 =pd.read_csv("Telcom_dataset.csv")
data_frame2.head(10)
```

```
#Loading second dataset
#
data_frame3 =pd.read_csv("Telcom_dataset2.csv")
data_frame3.head(10)
```

```
#Loading third dataset
#
data_frame4 =pd.read_csv("Telcom_dataset3.csv")
data_frame4.head(10)
```

**2.2.CLEANING DATAFRAMES**

**2.2.1.Renaming columns.**

```
#Telcom dataset
#
data_frame2= data_frame2.rename(columns={'PRODUTC':'PRODUCT','DATETIME':'DATE_TIME'})
data_frame2
```

```
#Telcom3 dataset
#
data_frame4= data_frame4.rename(columns={'CELLID':'CELL_ID','SIET_ID':'SITE_ID'})
data_frame4
```

**2.2.2.Duplicate Records**

```
#Telcom dataset
#
data_frame2.duplicated()
```

```
#Telcom2 dataset
#
data_frame3.duplicated()
```

```
#Telcom3 dataset
#
data_frame4.duplicated()
```

**2.2.3.Dropping duplicate records**

```
#Telcom dataset
#
data_frame2.drop_duplicates(inplace=True)
data_frame2
```

```
#Telcom2 dataset
#
data_frame4.drop_duplicates(inplace=True)
data_frame4
```

```
#Telcom3 dataset
#
data_frame4.drop_duplicates(inplace=True)
data_frame4
```

**2.3.Merging the DataFrames**

```
#Merging Telcom dataset and
#Telcom2 dataset
#
df1=pd.merge(data_frame2,data_frame3,how='outer',on=['PRODUCT','VALUE','DATE_TIME','CELL_ID','SITE_ID'])
df1
```

```
#Merging Telcom/Telcom2  and
#Telcom3 dataset
#
data_frame_telcom=pd.merge(df1,data_frame3,how='outer',on=['PRODUCT','VALUE','DATE_TIME','CELL_ID','SITE_ID'])
data_frame_telcom
```

**2.3.1.Merged DataFrame Cleaning**

```
#dropping unwanted columns
#
data_frame_telcom.drop(['COUNTRY_A','COUNTRY_B','COUNTRY_A_x','COUNTRY_A_y','COUNTRY_B_x','COUNTRY_B_y','CELL_ON_SITE','CELL_ON_SITE_x','CELL_ON_SITE_y','DW_A_NUMBER_INT','DW_B_NUMBER_INT','DW_A_NUMBER_x','DW_A_NUMBER_y','DW_B_NUMBER_x','DW_B_NUMBER_y'],axis=1,inplace=True)
data_frame_telcom
```

```
#checking for duplicates
#
data_frame_telcom.duplicated()
```

```
#dropping the duplicates
#
data_frame_telcom.drop_duplicates(inplace=True)
data_frame_telcom
```

```
#information about merged table
#
data_frame_telcom.info()
```

**3.1.Merging all dataframes**

```
#Merging
#
data_frame_combined=pd.merge(data_frame1,data_frame_telcom,how='left',on=['SITE_ID'])
data_frame_combined
```

```
#dropping unwanted columns
#
data_frame_combined.drop(['INDEX','STATUS','DECOUPZONE','LONGITUDE','LATITUDE','REGION','AREA','CELL_ID_x','CELL_ID_y'],axis=1,inplace=True)
data_frame_combined
```

**3.1.Merged DataFrame Cleaning**

```
#checking for duplicates
#
data_frame_combined.duplicated()
```

```
#dropping the duplicates
#
data_frame_combined.drop_duplicates(inplace=True)
data_frame_combined
```

```
#dropping null values
#
data_frame_combined.dropna(inplace=True)
data_frame_combined.head(5)
```

```
#resetting index
#
data_frame_combined.reset_index(inplace=True)
data_frame_combined
```

```
#dropping unwanted column
#
data_frame_combined.pop('index')
data_frame_combined
```

**4.DATA ANALYSIS**

**4.1.CREATING PRIORITY LISTS**

```
#creating main priority dataframe
#checking product used mostly
#
priority_data_frame1=pd.DataFrame(data_frame_combined.groupby(['VILLES','LOCALISATION','ZONENAME','SITE_ID','PRODUCT'])['VALUE'].sum())
priority_data_frame1
```

```
#cities with most 
#activity
#
priority_data_frame2= pd.pivot_table(priority_data_frame1, index=['VILLES'], aggfunc='sum')
priority_data_frame2
```

```
#sorting cities with most
#activities from the highest
#
priority_data_frame3=priority_data_frame2.sort_values('VALUE',ascending=False)
priority_data_frame3
```

```
#zonenames with most activity
#
priority_data_frame4= pd.pivot_table(priority_data_frame1, index=['ZONENAME','VILLES'],aggfunc='sum')
priority_data_frame4
```

```
#sorting in ascending order
#
priority_data_frame5=priority_data_frame4.sort_values('VALUE',ascending=False)
priority_data_frame5
```

```
#localisations with most activity
#
priority_data_frame6= pd.pivot_table(priority_data_frame1, index=['LOCALISATION'],aggfunc='sum')
priority_data_frame6
```
